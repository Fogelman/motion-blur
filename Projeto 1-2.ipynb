Para facilitar a experimentação:

- O Scikit-Image já tem uma função que faz o filtro de Wiener: ``skimage.restoration.wiener``

- Segue código para fazer um *kernel* de *motion blur* dado o tamanho do kernel, o comprimento do segmento de reta, e o ângulo.

```python
#cria o kernel a partir do angulo e tamanho
def make_psf(shape, length, angle):
    psf = np.zeros(shape)
    rows, cols = shape
    center = (cols//2, rows//2)
    psf[center[1], center[0]:(center[0]+length)] = 1.0
    M = cv2.getRotationMatrix2D(center, angle, 1)
    psf = cv2.warpAffine(psf, M, (cols, rows))
    psf = psf / psf.sum()
    return psf

def show(img):
    plt.figure(figsize=(8, 8))
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.show()
    
```

```python
def display_fft(img_freq, angle):
    img_freq_aux_mag = np.abs(img_freq)
    img_freq_aux_mag[0,0] = 0.0
    img_freq_aux_mag_display = np.log10(1 + np.fft.fftshift(img_freq_aux_mag))
    plt.figure(figsize=(8,8))
    
    row, col = img_freq.shape
    x, y = [col//2,0 ], [row//2, 243]
    print(f"angle: {np.arctan((y[1] - y[0])/(x[1] - x[0]))*(180/np.pi)}")
    
    plt.imshow(img_freq_aux_mag_display, cmap='gray')
    plt.plot(x, y, marker = 'o')
    plt.axis('off')
    plt.show()
    
    
    
```

# FFT

Para determinar os parametros do kernel do motion blur, foi aplicado o fft

Com isso foi possivel determinar o angulo que as faixas brancas estão orientadas. A linha azul representa a direção das faixas

```python
# forest glen 1200
img_test_freq = np.fft.fft2(img_blur)
display_fft(img_test_freq, 12.5525)
```

```python
kernel = np.fft.fft2(make_psf(img_blur.shape, 20,167.46))
display_fft(kernel, 12.5525)
```

```python
display_fft(img_test_freq-kernel, 12)
```

```python
def wiener(G, H, K):
    W = H.conj() / (np.abs(H)**2 + K)
    F = W*G
    return F


# Chuta o K.
K = 0.001

h = make_psf(giant_blur_img.shape, 20,167.46)
H = np.fft.fft2(h)

# Filtra.
G = np.fft.fft2(giant_blur_img)
img_test_blur_recon_wiener_freq = wiener(G, H, K)
img_test_blur_recon_wiener = np.fft.fftshift(np.fft.ifft2(img_test_blur_recon_wiener_freq).real)

plt.figure(figsize=(8,8))
plt.imshow(np.clip(img_test_blur_recon_wiener[:rows, :cols], 0, 1), cmap='gray')
plt.axis('off')
plt.show()
```

```python
img_deblured = img_test_blur_recon_wiener[:rows, :cols].copy()
```

## Clahe

```python
clahe = cv2.createCLAHE(clipLimit=0.2, tileGridSize=(15,15))
img_clahe = clahe.apply((img_deblured*255).astype(np.uint8))
show(img_clahe)
```

```python
img_perspective = img_clahe / 255.0
rows, cols = (256,1024)

pts_original = np.float32([[400,230],[640,170],[400,300],[640,235]])
pts_corrigida = np.float32([[0, 0], [cols - 1, 0], [0, rows - 1], [cols - 1, rows - 1]])

M = cv2.getPerspectiveTransform(pts_original, pts_corrigida)

img_corrigida = cv2.warpPerspective(img_perspective, M, (cols, rows))

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(img_perspective, cmap='gray')
for k, p in enumerate(pts_original):
    plt.plot([p[0]], [p[1]], 'ro')
    plt.text(p[0] + 5, p[1], str(k), color='red')

plt.subplot(122)
plt.imshow(img_corrigida, cmap='gray')
for k, p in enumerate(pts_corrigida):
    plt.plot([p[0]], [p[1]], 'ro')
    plt.text(p[0] + 5, p[1], str(k), color='red')
plt.show()
```
